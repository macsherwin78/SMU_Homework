```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
unique_player_count = len(purchase_data["SN"].unique())

player_count_df = pd.DataFrame([{"Total Players": unique_player_count}])
player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item Name"].unique())

Price_sum = purchase_data["Price"].sum()

num_Purchase = len(purchase_data["Purchase ID"])

Ave_Price = Price_sum/num_Purchase

Purch_Anlyss_df = pd.DataFrame([
    {"Number of Unique Items": unique_items,
     "Average Price": Ave_Price,
     "Number of Purchases": num_Purchase,
     "Total Revenue": Price_sum}])

Purch_Anlyss_df['Average Price'] = Purch_Anlyss_df['Average Price'].astype(float).map("${:,.2f}".format)
Purch_Anlyss_df['Total Revenue'] = Purch_Anlyss_df['Total Revenue'].astype(float).map("${:,.2f}".format)
Purch_Anlyss_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
unique_SN_df = purchase_data.drop_duplicates("SN")

count_by_gender = pd.DataFrame(unique_SN_df["Gender"].value_counts())
count_by_gender = count_by_gender.rename(columns={"Gender": "Total Count"})

prcnt_plyrs = pd.DataFrame(unique_SN_df["Gender"].value_counts()/unique_player_count)
prcnt_plyrs = prcnt_plyrs.rename(columns={"Gender": "Percentage of Players"})

count_by_gender['Percentage of Players'] = prcnt_plyrs

count_by_gender['Percentage of Players'] = count_by_gender['Percentage of Players'].astype(float).map("{:,.2%}".format)
count_by_gender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purch_count_byGndr = pd.DataFrame(purchase_data["Gender"].value_counts())
purch_count_byGndr_df = purch_count_byGndr.rename(columns={"Gender": "Purchase Count"})

purchase_data_Gender = purchase_data.groupby(["Gender"])

total_purchVal_byGndr = pd.DataFrame(purchase_data_Gender.Price.sum())
purch_count_byGndr_df['Total Purchase Value'] = total_purchVal_byGndr

purch_count_byGndr_df['Average Purchase Price'] = purchase_data_Gender.Price.sum()/purchase_data["Gender"].value_counts()

purch_count_byGndr_df['Avg Total Purchase per Person'] = purchase_data_Gender.Price.sum() / unique_SN_df["Gender"].value_counts()
```

```python
purch_count_byGndr_df["Total Purchase Value"] = purch_count_byGndr_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purch_count_byGndr_df["Average Purchase Price"] = purch_count_byGndr_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purch_count_byGndr_df["Avg Total Purchase per Person"] = purch_count_byGndr_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

purch_count_byGndr_df

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9.999, 14.999, 19.999, 24.999, 29.999, 34.999, 39.999, 100]

age_group_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

pd.cut(unique_SN_df["Age"], age_bins, labels=age_group_labels)

unique_SN_df = pd.DataFrame(unique_SN_df)
unique_SN_df["Age_Group"] = pd.cut(unique_SN_df["Age"], age_bins, labels=age_group_labels)

Age_Group_df = pd.DataFrame(unique_SN_df.Age_Group.value_counts())
Age_Group_df = Age_Group_df.rename(columns={"Age_Group": "Total Count"})

Age_Group_df["Percentage of Players"] = unique_SN_df.Age_Group.value_counts()/unique_player_count

Age_Group_df["Percentage of Players"] = Age_Group_df["Percentage of Players"].astype(float).map("{:,.2%}".format)

Age_Group_df.sort_index(axis=0)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins = [0, 9.999, 14.999, 19.999, 24.999, 29.999, 34.999, 39.999, 100]

age_group_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

purchase_data["Age_Group"] = pd.cut(purchase_data["Age"], age_bins, labels=age_group_labels)

purchase_count_byAge = pd.DataFrame(purchase_data.Age_Group.value_counts())
purchase_count_byAge = purchase_count_byAge.sort_index(axis=0)
purchase_count_byAge = purchase_count_byAge.rename(columns={"Age_Group": "Purchase Count"})

purchase_data_Age_df = purchase_data.groupby(["Age_Group"])

total_purchVal_byAge = pd.DataFrame(purchase_data_Age_df.Price.sum())
purchase_count_byAge['Total Purchase Value'] = total_purchVal_byAge

ave_Price_byAge = purchase_data_Age_df.Price.sum()/purchase_data.Age_Group.value_counts()
purchase_count_byAge['Average Purchase Price'] = ave_Price_byAge

ave_TotalPur_byAge = purchase_data_Age_df.Price.sum()/unique_SN_df.Age_Group.value_counts()
purchase_count_byAge['Avg Total Purchase per Person'] = ave_TotalPur_byAge
```

```python
purchase_count_byAge["Average Purchase Price"] = purchase_count_byAge["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchase_count_byAge["Total Purchase Value"] = purchase_count_byAge["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purchase_count_byAge["Avg Total Purchase per Person"] = purchase_count_byAge["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
purchase_count_byAge
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Extract item Data
item_data = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

# Perform basic calculations
total_item_purchase = item_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
average_item_purchase = item_data.groupby(["Item ID", "Item Name"]).mean()["Price"]
item_count = item_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")

item_data_df = pd.DataFrame(item_count)
item_data_df['Item Price']=average_item_purchase
item_data_df['Total Purchase Value']=total_item_purchase

item_data_df["Item Price"] = item_data_df["Item Price"].astype(float).map("${:,.2f}".format)
item_data_df["Total Purchase Value"] = item_data_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)

item_data_df.sort_values('Purchase Count', ascending=False).head(5)
```
